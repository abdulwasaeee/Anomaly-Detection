```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.neighbors import LocalOutlierFactor
import  matplotlib.pyplot as plt
```

```python
x, _ = make_blobs(n_samples=300, centers=2, cluster_std=1.0, random_state=42)
```

```python
n_outliers = 10
rng = np.random.RandomState(42)
outliers = rng.uniform(low=-10, high=10, size=(n_outliers, 2))
x = np.vstack([x, outliers])
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.03)
y_pred = lof.fit_predict(x)
outliers=np.where(y_pred<0)
```

```python
plt.scatter(x[:, 0], x[:, 1], label="Inliers")
plt.scatter(x[outliers, 0], x[outliers, 1], c='red', label="Outliers")
plt.legend()
```
